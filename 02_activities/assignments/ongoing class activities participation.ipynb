```python
# To make it a line plot
np.random.seed(123)
x = np.arange(50)
y =np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# To add axis labels and title
np.random.seed(123)
x = np.arange(50)
y =np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

```python
# Modifying labels and titles by using font dictionaries

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total Growth', fontdict = font2)
ax.set_xlabel('Years since start')
```

```python
# To center align the title #Moving labels and titles with loc argument
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total Growth', fontdict = font2)
ax.set_xlabel('Years since start')
```

```python
# customizing plot appearance #Modifying data points
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y, marker = '*', #New additions
           color ='indigo') #New additions
fig.show()
```

```python
# Adjust the style and width of the line connecting our points
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = 'v', 
            color ='indigo',
            linestyle ='--', # New addition
            linewidth= 2) # New addition
fig.show
```

```python
# Modifying data points: Color
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = 'o',
            markersize=12,
            color ='#7425b9',
            linestyle ='--',
            linewidth= 2,
            )
```

```python
# Modifying data points: Color. Customize our plot further
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = '.',
            markersize=12,
            color ='#7425b9',
            linestyle ='--',
            linewidth= 2,
            markeredgecolor = '#fa9359',
            markerfacecolor ='#000000'
            )
```

```python
# To add gridlines to axis
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = 'D', 
            markersize=12,
            color ='indigo',
            linestyle ='--',
            linewidth= 2,
            markeredgecolor = '#fa9359',
            markerfacecolor ='#000000')
ax.grid(axis = 'y')
```

```python
# Modify the appearance (color, width, linestyle) of our grid lines
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = 'H', 
            markersize=12,
            color ='indigo',
            linestyle ='--',
            linewidth= 2,
            markeredgecolor = '#fa9359',
            markerfacecolor ='#000000')
ax.grid(axis='y', color='green')
```

```python
# Exact code copied from seaborn barplot example
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set plot style: grey grid in the background:
sns.set_theme(style="darkgrid")

# load dataset
tips = sns.load_dataset("tips")

# Set the figure size
plt.figure(figsize=(10, 7))

# plot a bar chart
sns.barplot(
    x="total_bill", 
    y="day", 
    data=tips, 
    estimator=sum, 
    ci=None, 
    color='#69b3a2');
plt.title('Total Bill Amount per Day', fontsize=16)
```

```python
# Modified seaborn barplot with customizations
# Set plot style
sns.set_theme(style="whitegrid")  # Changed from darkgrid to whitegrid

# Load dataset
tips = sns.load_dataset("tips")

# Set figure size
plt.figure(figsize=(10, 7))

# Create a customized barplot
sns.barplot(
    x="total_bill",      # Keep total bill on x-axis
    y="day",             # Days on y-axis
    data=tips,
    estimator=np.mean,   # Show average instead of sum
    ci=None,
    palette="Set2"       # Use a custom color palette
)

# Add title and labels
plt.title("Average Total Bill by Day of the Week", fontsize=16, weight='bold')
plt.xlabel("Average Total Bill ($)", fontsize=12)
plt.ylabel("Day of the Week", fontsize=12)

# add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)
```

```python
# Adding annotations with the text() function
# Start by putting our data into a scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label ="Person 1")
ax.scatter(x,y2, label ="Person 2")
ax.legend(loc ='lower right')
# Use the text function to specifty coordinates and text
ax.text(10, 95, "This value is important!")
```

```python
# Modifying text annotations by adding arguments
# Start by putting our data into a scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label ="Person 1")
ax.scatter(x,y2, label ="Person 2")
ax.legend(loc ='lower right')
# Use the text function to specifty coordinates and text
ax.text(10, 95, "This value is important!",
ha = 'center', #alignment of text
color = 'red', # modify font color
size = 20) # modify font size
```

Axis Labels

```python
# Removing tick marks and labels
# Use locator to remove ticks and labels
# Use formatter to remove only labels

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label= "Person 2")
ax.legend(loc='lower right')


ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Limiting tick marks
# we can define a maximum number of ticks we want on our axis
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label= "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Tick Mark intervals
# We can set tick marks at a particular interval (in this case, 5) using MultipleLocator
# set major locator to multiples of 5 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label= "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotating axis labels
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label= "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right')
```

```python
# Modify figure layout
# Adding some very large x axis titles to our previous plot
fig, someaxes = plt.subplot_mosaic ([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1) 

someaxes ["ax1"].set_xlabel('A big Label', fontsize=18)
someaxes ["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes ["ax3"].set_xlabel('Label 2: 2  Fast 2 Furious', fontsize=18)



#Result shows that the axis title for our scatter plot doesn't appear
# The axis title for our line plot is cut off
```

```python
# Layouts
# We can use layouts to make sure that our subplots fit neatly in our figure area
# Two Main kinds of layour to consider:
# Tight layout: Makes sure that the tick labels, axis labels doesn't cut off as seen above 
# Constrained layout: works similarly except it also fit things like legends or colorbars

#Using constrained layout on our plot

fig, someaxes = plt.subplot_mosaic ([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4),
                                 layout = "constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1) 
someaxes ["ax1"].set_xlabel('A big Label', fontsize=18)
someaxes ["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes ["ax3"].set_xlabel('Label 2: 2  Fast 2 Furious', fontsize=18)
```

Multple viz on one axes object

```python
# Make our sample data
x= np.array (["Luffy","Zoro","Nami","Usopp","Sanji"])
y1=np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])

# Define figure and axes object
fig, ax = plt.subplots(figsize =(7,3))
# Now call both bar and plot elements to the same axes
ax.bar(x,y1,color = "indigo")
ax.plot(x,y2,color = "red")
```

```python
# Add error information
# First, calculate standard deviation of our data
#std on y2
y2_sd = np.std(y2)

# Then plot our line as before
fig, ax = plt.subplots(figsize =(7,3))
ax.plot(x,y2, color = "red")
```

```python
#Then use errorbar() to add in our error line
#fmt means format
# Add errorbar on the standard deviation calculated earlier
fig, ax = plt.subplots(figsize =(7,3))
ax.plot(x,y2, color = "red")
ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none")
```

```python
# Customizing errobar appearance
fig, ax = plt.subplots(figsize =(7,3))
ax.plot(x,y2, 
        color = "red")
ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick=4)
```

```python
"""""
If we don;t want to see error bars for every single point, we can 
specify intervals using errorevery
"""""
fig, ax = plt.subplots(figsize =(7,3))
ax.plot(x,y2, 
        color = "red")
ax.errorbar(x,
            y2,
            yerr = y2_sd,
            fmt="none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick=4,
            errorevery=2
            )
```
